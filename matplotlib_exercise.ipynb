## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv('data\hist_data.csv')
plt.title("aデータとbデータの温度に関する度数分布", fontsize=10) 
plt.xlabel("温度(℃)", fontsize=10)
plt.ylabel("度数", fontsize=10)
plt.hist(df_hist['aデータ'], alpha=0.5, bins=30, label='aデータ', color='gold')
plt.hist(df_hist['bデータ'], alpha=0.5, bins=30, label='bデータ', color='cyan')
plt.legend(loc="upper right", fontsize=10)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv('data\scatter_data.csv')
df_scat0 = df_scatter[df_scatter['class'] == 0]
df_scat1 = df_scatter[df_scatter['class'] == 1]
df_scat2 = df_scatter[df_scatter['class'] == 2]

plt.title("クラス０～２のlengthとwidthに関する散布図", fontsize=10) 
plt.xlabel("length(cm)", fontsize=10)
plt.ylabel("width(cm)", fontsize=10)
plt.grid()
plt.scatter(df_scat0['length (cm)'], df_scat0['width (cm)'], label='class0', color='red')
plt.scatter(df_scat1['length (cm)'], df_scat1['width (cm)'], label='class1', color='blue')
plt.scatter(df_scat2['length (cm)'], df_scat2['width (cm)'], label='class2', color='green')
plt.legend(loc="upper right", fontsize=10)

```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv('data\line_data.csv')
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)

ax1.plot(df_line['期間'], df_line['件数'], marker="o", markeredgewidth=0)
plt.title("期間ごとの件数の推移", fontsize=10) 
plt.xlabel("期間", fontsize=10)
plt.ylabel("件数", fontsize=10)
plt.grid()
plt.xticks(rotation=30)

ax2 = fig.add_subplot(2, 1, 2)
ax2.bar(df_line['期間'], df_line['費用'])
plt.title("期間ごとの費用の推移", fontsize=10) 
plt.xlabel("期間", fontsize=10)
plt.ylabel("費用(円)", fontsize=10)
plt.grid()
plt.xticks(rotation=30)

plt.subplots_adjust(top=2, hspace=0.4)
```
